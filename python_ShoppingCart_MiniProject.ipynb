```python
import pandas as pd
dict1= {'Sr.no' : [1,2,3],
        'Item' : ['Biscuits','Cereals','Chicken'],
        'Quantity' : [5,10,20],
        'Cost/Item' : [20,90,100]
       }
SC = pd.DataFrame(dict1)
```

```python
SC
```

```python
bill_columns = ['S.no', 'Item', 'Quantity', 'Total Cost']
bill_df = pd.DataFrame(columns=bill_columns)
count = 0
while (True):
    user_input = int(input("What do you want to purchase? 1.Biscuts,2.cereals,3.chicken"))
    if (user_input == 1):
        input2 = int(input("How many Biscuits packets you want to purchase:"))
        Available_Quantity = SC.loc[SC['Item'] == 'Biscuits', 'Quantity'].iloc[0]
        if (input2 > Available_Quantity):
            print("Available quantity of Biscuits is 5:")
        elif(input2 < 0):
            print("entered invalid quantity")
        elif (input2 <= Available_Quantity):
            # code for item quantity
            SC.loc[SC['Item'] == 'Biscuits', 'Quantity'] -= input2
            total_cost = input2 * SC.loc[SC['Item'] == 'Biscuits', 'Cost/Item'].values[0]
            count += 1
            bill_df = bill_df._append({
                'S.no': len(bill_df) + 1,
                'Item': "Biscuits",
                'Quantity': input2,
                'Total Cost': total_cost
            }, ignore_index=True)
        input3 = input("Do you want to continue shopping? Y/N")
        if (input3 == 'N'):
            if(count > 0):
                input4 = input("Enter your name : ")
                input5 = input("Enter your adress : ")
                input6 = int(input("Enter the distance from store 5/10/15/30 : "))
                if (input6 > 0 and input6 <= 15):
                    dc = 50
                    print('Delivery charge: 50Rs will be levied for distance between 0 to 15')
                elif (input6 >15 and input6 <= 30):
                    dc = 100
                    print('Delivery charge: 100Rps will be levied for distance between 0 to 15')
                print("Generating Bill...\n")
                print(bill_df)
                print("Total items cost:" ,bill_df['Total Cost'].sum())
                print("Total Bill Amount:total items cost + Delivery charges",bill_df['Total Cost'].sum()+dc)
                print("Name:"+input4)
                print("Address:"+input5)
                print("Thank you for shopping with us!")
            break
    elif (user_input == 2):
        input2 = int(input("How many Cereal packets you want to purchase:"))
        Available_Quantity = SC.loc[SC['Item'] == 'Cereals', 'Quantity'].iloc[0]
        if (input2 > Available_Quantity):
            print("Available quantity of cereals is 10:")
        elif(input2 < 0):
            print("entered invalid quantity")
        elif (input2 <= Available_Quantity):
            # code for item quantity
            SC.loc[SC['Item'] == 'Cereals', 'Quantity'] -= input2
            total_cost = input2 * SC.loc[SC['Item'] == 'Cereals', 'Cost/Item'].values[0]
            count += 1
            bill_df = bill_df._append({
                'S.no': len(bill_df) + 1,
                'Item': 'Cereals',
                'Quantity': input2,
                'Total Cost': total_cost
            }, ignore_index=True)
        input3 = input("Do you want to continue shopping? Y/N")
        if (input3 == 'N'):
            if(count > 0):
                input4 = input("Enter your name : ")
                input5 = input("Enter your adress : ")
                input6 = int(input("Enter the distance from store 5/10/15/30 : "))
                if (input6 > 0 and input6 <= 15):
                    dc = 50
                    print('Delivery charge: 50Rs will be levied for distance between 0 to 15')
                elif (input6 >15 and input6 <= 30):
                    dc = 100
                    print('Delivery charge: 100Rps will be levied for distance between 0 to 15')
                print("Generating Bill...\n")
                print(bill_df)
                print("Total items cost:" ,bill_df['Total Cost'].sum())
                print("Total Bill Amount:total items cost + Delivery charges",bill_df['Total Cost'].sum()+dc)
                print("Name:"+input4)
                print("Address:"+input5)
                
                print("Thank you for shopping with us!")
            break
    elif (user_input == 3):
        input2 = int(input("How many Chicken packets you want to purchase:"))
        Available_Quantity = SC.loc[SC['Item'] == 'Chicken', 'Quantity'].iloc[0]

        if (input2 > Available_Quantity):
            print("Available quantity of chicken is 20:")
        elif(input2 < 0):
            print("entered invalid quantity")
        elif (input2 <= Available_Quantity):
            # code for item quantity
            SC.loc[SC['Item'] == 'Chicken', 'Quantity'] -= input2
            total_cost = input2 * SC.loc[SC['Item'] == 'Chicken', 'Cost/Item'].values[0]
            count += 1
            bill_df = bill_df._append({
                'S.no': len(bill_df) + 1,
                'Item': 'Chicken',
                'Quantity': input2,
                'Total Cost': total_cost
            }, ignore_index=True)
        input3 = input("Do you want to continue shopping? Y/N")
        if (input3 == 'N'):
            if(count > 0):
                input4 = input("Enter your name : ")
                input5 = input("Enter your adress : ")
                input6 = int(input("Enter the distance from store 5/10/15/30 : "))
                if (input6 > 0 and input6 <= 15):
                    dc = 50
                    print('Delivery charge: 50Rs will be levied for distance between 0 to 15')
                elif (input6 >15 and input6 <= 30):
                    dc = 100
                    print('Delivery charge: 100Rps will be levied for distance between 0 to 15')
                print("Generating Bill...\n")
                print(bill_df)
                print("Total items cost:" ,bill_df['Total Cost'].sum())
                print("Total Bill Amount:total items cost + Delivery charges",bill_df['Total Cost'].sum()+dc)
                print("Name:"+input4)
                print("Address:"+input5)
                print("Thank you for shopping with us!")
                break
    else:
        print("invalid item selected")
        
   
```
