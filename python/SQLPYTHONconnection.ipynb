```python
import mysql.connector as conn
import pandas as pd
```

```python
mydb = conn.connect(host="localhost", user="root", passwd="root",auth_plugin='mysql_native_password')
                             # auth_plugin='mysql_native_password')#,auth_plugin='mysql_native_password')
```

```python
mydb
```

```python
cursor = mydb.cursor()
```

```python
#Creating Database in local machine
cursor.execute("create database ShoppingCart_DB")
```

```python
#Creating table in database
connection = cursor.execute("""create table shoppingcart_DB.listofitems(
    srno int, 
    Item varchar(15),
    Quantity int,
    cost int)
    """)
```

```python
# Define the SQL query to insert values into the table
insert_query = """
INSERT INTO ShoppingCart_DB.listofitems (srno, Item, Quantity, Cost)
VALUES (%s, %s, %s, %s)
"""
```

```python
# Sample values to insert
values_to_insert = [
    (1, 'Biscuits', 5, 20),
    (2, 'Cereals', 10, 90),
    (3, 'Chicken', 20, 100)
]
```

```python
cursor.executemany(insert_query, values_to_insert)
```

```python
mydb.commit()
```

```python
inventory = pd.read_sql('select * from ShoppingCart_DB.listofitems', mydb)
```
